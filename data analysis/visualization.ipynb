Graph Types
- Line Graph : Show trends over time
- Bar Graph : Compare values between categories(discreate data)
- Pie Chart : compare values between categories (percentage data)
- Histogram : Show distribution of numerical data 
- Box Plot : statistical distribution of numerical data
- Heat Map : show matrix of values using color

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
fig,ax = plt.subplots(figsize = (12,4))
ax.plot('total_bill',data = tips)
```

```python
fig,ax = plt.subplots(figsize = (12,4))
ax.plot('total_bill',data = tips,color='red',marker='.')
```

```python
fig,ax = plt.subplots(figsize = (12,4))
ax.plot(tips.index,'total_bill',data = tips,color='red')
```

```python
fig,ax = plt.subplots(figsize = (20,4))
ax.plot(tips.index,'total_bill',data = tips,color='red')
ax.plot(tips.index,'tip',data = tips,color='green')
ax.set_xlim(0,100)
```

```python
fig,ax = plt.subplots(figsize = (20,4))
ax.bar(tips.index,'total_bill',data = tips,color='red')
ax.bar(tips.index,'tip',data = tips,color='green')
ax.set_xlim(0,100)
```

```python
fig,ax = plt.subplots(figsize = (20,4))
plt.stem('total_bill',data=tips,linefmt='red',markerfmt='D')
```
