```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv("spotify.csv", encoding='latin1')
df.head()
```

```python
df.columns.tolist()
```

```python
df.sort_values(by='artist_count', ascending=False).head(25)
```

```python
df.sort_values(by='artist_count',ascending=False).drop_duplicates(subset='track_name')
```

```python
df.sort_values(by='artist_count', ascending=False).head(25).drop_duplicates(subset='artist(s)_name')
```

```python
top_10_track = df.sort_values(by='artist_count', ascending=False).head(25).drop_duplicates(subset='track_name')[['track_name', 'artist_count']]
px.bar(top_10_track, 'artist_count', 'track_name', title='Top 10 Tracks')
```

```python
released_year = df.groupby('released_year')
for year in df.released_year.unique().tolist():
    sdf = released_year.get_group(year)
    player_df = sdf.sort_values(by='artist_count', ascending=False).head()[['artist(s)_name', 'artist_count']]
    fig = px.bar(player_df, 'artist(s)_name','artist_count', title=f'Top 5 Track in {released_year}',width=500)
    fig.show()
```
